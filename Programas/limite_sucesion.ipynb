```python
import matplotlib.pyplot as plt
import numpy as np
import random

import os

colores =   [
    "red", "blue", "green", "orange", "yellow",
    "purple", "pink", "brown", "gray", "black",
    "white", "cyan", "magenta",
    "gold", "silver", "turquoise", "violet", "indigo",
    "lime", "olive", "teal", "coral", "orchid",
    "salmon", "maroon", "navy", "sienna", "wheat",
    "darkred", "darkblue", "darkgreen", "darkorange", "darkviolet",
    "deeppink", "chocolate", "slategray", "midnightblue", "peru",
    "aquamarine", "cadetblue", "crimson", "darkcyan", "darkolivegreen",
    "firebrick", "forestgreen", "hotpink", "khaki", "lightcoral",
    "mediumaquamarine", "mediumseagreen", "mediumvioletred", "olivedrab", "orangered",
    "palegreen", "paleturquoise", "palevioletred", "royalblue", "saddlebrown",
    "seagreen", "skyblue", "springgreen", "steelblue", "tomato"
]

a = 0; b = 10 #Aqu se define el intervalo

# Enseguida se define la funcin f(x) mediante su regla de correspondencia:
def f(x):
    fun = '1 - 8*((-1)/2)**x'
    return eval(fun)

x = np.arange(1,b+1) 
y = f(x) 

plt.figure(figsize=(12,8),dpi=1200)


minejey=-4
maxejey=6
minejex=0
maxejex=b

plt.arrow(0, minejey,0,maxejey-minejey, head_width=0.1, head_length=0.1, fc='black', ec='black',linewidth=1)
plt.arrow(-1, minejex,maxejex+1, 0, head_width=0.1, head_length=0.1, fc='black', ec='black',linewidth=1)

# labels=[]
# for i in range(0,k):
#     labels.append("a" +str(i))

xlabel=list(range(minejex,maxejex))

 
for i in range(minejex,maxejex):    
        plt.text(i,-0.2, xlabel[i], fontsize=10, ha='center', va='bottom')
# for i in range(minejey,maxejey):
#         if i != 0:    
#                 plt.text(-0.15,i-0.15, str(i), fontsize=10, ha='center', va='bottom')
for i in range(len(x)-1):
        plt.plot(x[i],y[i],color=random.choice(colores), marker = 'o',markersize = 2.5)
        #plt.text(x[i], y[i], labels[i], fontsize=10, ha='center', va='bottom')
    
plt.axhline(y=1,xmin=0,xmax=12, color='r', linestyle='-', linewidth=0.5)
plt.text(-0.1,1, '$L$', fontsize=10, ha='center', va='bottom')
plt.axhline(y=1-0.35,xmin=0,xmax=12, color='b', linestyle='--', linewidth=0.35)
plt.text(-0.25,1+0.35, '$L+ \epsilon$', fontsize=10, ha='center', va='bottom')
plt.axhline(y=1+0.35,xmin=0,xmax=12, color='b', linestyle='--', linewidth=0.35)
plt.text(-0.25,1-0.55, '$L- \epsilon$', fontsize=10, ha='center', va='bottom')
plt.yticks([]) 
plt.xticks([]) 
plt.savefig('limite1.png')
plt.show()
```
