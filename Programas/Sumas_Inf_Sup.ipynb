```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

# Parámetros
a = 0
b = 5
n = 20

# Función continua en [a, b]
def f(x):
    return math.sqrt(x)

# Generar partición uniforme
#x_partition = np.linspace(a, b, n + 1)
internal_points = [random.uniform(a, b) for _ in range(n - 1)]
x_partition = sorted([a] + internal_points + [b])  # x_0 = a, x_n = b

# Generar puntos x_i^* aleatorios en cada subintervalo (x_{i-1}, x_i)
x_stars = [random.uniform(x_partition[i], x_partition[i + 1]) for i in range(n)]
heights = [f(x_star) for x_star in x_stars]

# Crear figura
plt.figure(figsize=(10, 5))

# Graficar los rectángulos estilo escalonado
for i in range(n):
    xi_left = x_partition[i]
    xi_right = x_partition[i + 1]
    yi = heights[i]

    # Tramo horizontal azul grueso
    plt.plot([xi_left, xi_right], [yi, yi], color='blue', linewidth=1)

    # Relleno celeste semitransparente
    plt.fill_between([xi_left, xi_right], [yi, yi], color='lightblue', alpha=0.4)

    # Líneas verticales punteadas grises
    plt.vlines(xi_left, ymin=0, ymax=yi, linestyle='--', color='gray', linewidth=1)
    plt.vlines(xi_right, ymin=0, ymax=yi, linestyle='--', color='gray', linewidth=1)

    # Puntos extremos
    plt.plot(xi_left, yi, marker='o', color='blue', markerfacecolor='white', linewidth=1)
    plt.plot(xi_right, yi, marker='o', color='blue', markerfacecolor='white', linewidth=1)  # Punto abierto

# Último punto (opcional)
plt.plot(x_partition[-1], heights[-1], color='blue', markerfacecolor='white', linewidth=1)

# Graficar la curva f(x)
x_vals = np.linspace(a, b, 500)
y_vals = [f(x) for x in x_vals]
plt.plot(x_vals, y_vals, color='black', label='$f(x) = sqrt{x}$')

# Ejes y detalles
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sumas de Riemann')
plt.grid(True, linestyle='--', alpha=0.6)
# plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math 

np.random.seed(10)

a = -5; b = 5
k=1000
n=10

def f(x):
    fun = 'np.sin(x)**2'
    return eval(fun)

z = np.linspace(a,b,k)
y = f(z)

plt.axhline(0, linestyle='-', linewidth=1, color='black')
plt.axvline(0, linestyle='-', linewidth=1, color='black')

plt.plot(z,y, color = 'Tomato', linewidth = 1)

x = np.linspace(a,b,n)
sum = 0;
delta = (b-a)/n
k=10
for i  in range(len(x)-1):
    z=np.linspace(x[i],x[i+1],k)
    
    ci = max(np.array([f(w) for w in z]))
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'darkblue', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='blue', alpha=.5)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1 )
    sum += ci*delta 
    
plt.xlabel('x')
plt.ylabel('f(x)=x^2')
plt.title('Suma de Riemann')
plt.show()


print('La suma de Riemann es:',sum)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

# Parámetros
a, b = 0, 5
n = 10

# Función continua en [a, b]
def f(x):
    return math.sqrt(x)

# Generar partición aleatoria ordenada con puntos en (a,b)
internal_points = sorted(random.uniform(a, b) for _ in range(n - 1))
x_partition = [a] + internal_points + [b]

# Generar puntos x_i^* aleatorios dentro de cada subintervalo (x_{i-1}, x_i)
x_stars = [random.uniform(x_partition[i], x_partition[i + 1]) for i in range(n)]
heights = [f(x_star) for x_star in x_stars]

# Crear figura
plt.figure(figsize=(10, 8))
sum = 0
# Graficar los rectángulos estilo escalonado
for i in range(n):
    xi_left, xi_right = x_partition[i], x_partition[i + 1]
    yi = heights[i]
    sum = sum + (xi_right - xi_left)* yi
    # Tramo horizontal azul grueso
    plt.plot([xi_left, xi_right], [yi, yi], color='blue', linewidth=2)

    # Relleno celeste semitransparente
    plt.fill_between([xi_left, xi_right], [0, 0], [yi, yi], color='lightblue', alpha=0.4)

    # Líneas verticales punteadas grises
    plt.vlines([xi_left, xi_right], ymin=0, ymax=yi, linestyle='--', color='gray', linewidth=1)

    # Puntos extremos (círculos blancos con borde azul)
    plt.plot([xi_left, xi_right], [yi, yi], 'o', markerfacecolor='white', markeredgecolor='blue', markersize=4.5)

# Graficar la curva f(x)
x_vals = np.linspace(a, b, 500)
y_vals = [f(x) for x in x_vals]
plt.plot(x_vals, y_vals, color='black')

# Ejes y detalles
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Suma de Riemann de una partición aleatoria')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print('La suma de Riemann es:',sum)

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#np.random.seed(10)

a = 0; b = 5 #Aquí se define el intervalo

k=10000  #Número de puntos a gráficar
n=8 #Número de puntos en la particion del intervalo [a,b]

# Enseguida se define la función f(x) mediante su regla de correspondencia:
def f(x):
    fun = 'x**2'
    return eval(fun)

z = np.linspace(a,b,k) #Array con los puntos que se van a utilizar para gráficar
y = f(z) #Arrar con las imagenes de los puntos en z

#Se "dibujan" los ejes
plt.axhline(0, linestyle='-', linewidth=1, color='black')
plt.axvline(0, linestyle='-', linewidth=1, color='black')

#Se dibuja la grafica de la función f(x)  y se sombrea el area debajo de la gráfica
plt.fill_between(z,y,color='gray', alpha=0.5)
plt.plot(z,y, color = 'red', linewidth = 1.5)


x = np.linspace(a,b,n)  #Array que determina la partición homegenea de [a,b] de longitud n
delta = (b-a)/n #Longitud de cada uno de los subintervalos determinados por dos puntos consecutivos de la partición homogenea

#Enseguida inicializamos la suma inferior y superior en Cero
suma_inferior = 0;
suma_superior = 0;

#Graficamos los rectangulos inscritos
for i  in range(len(x)-1):
    ci = f(x[i])
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'darkblue', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='darkblue', alpha=0.2)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1.5 )
    suma_inferior += ci*delta

#Graficamos los rectangulos inscritos
for i  in range(len(x)-1):
    ci = f(x[i+1])
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'green', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='green', alpha=0.15)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1.5 )
    suma_superior += ci*delta

plt.xlabel('$x$')
plt.ylabel('$f(x)=x^2$')
#plt.title('Suma de Riemann')
plt.show()


print(f"La suma inferior de Riemann es: {suma_inferior}. ", end='')
print(' |  La suma superior de Riemann es:',suma_superior)

```

```python
import turtle

def koch_curve(t, order, size):
    if order == 0:
        t.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
            koch_curve(t, order - 1, size / 3)
            t.left(angle)

def draw_koch_curves():
    window = turtle.Screen()
    window.bgcolor("white")
    window.title("Centered Vertical Koch Curves")

    orders = [0, 1, 2, 3, 4,5]
    size = 500

    initial_x = -size   # Posición inicial en el eje x
    initial_y = size - 100   # Posición inicial en el eje y

    for order in orders:
        koch_turtle = turtle.Turtle()
        koch_turtle.speed(0)
        koch_turtle.hideturtle()
        koch_turtle.color("blue")

        koch_turtle.penup()
        koch_turtle.goto(initial_x, initial_y)
        koch_turtle.pendown()

        koch_curve(koch_turtle, order, size)

        initial_y -= size * 0.32 # Separación vertical entre copos

    turtle.done()

def main():
    draw_koch_curves()

if __name__ == "__main__":
    main()
```

```python
import turtle

def koch_snowflake(t, order, size):
    if order == 0:
        t.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
            koch_snowflake(t, order - 1, size / 3)
            t.left(angle)

def main():
    window = turtle.Screen()
    window.bgcolor("white")
    window.title("Koch Snowflake")

    snowflake = turtle.Turtle()
    snowflake.speed(0)
    snowflake.color("blue")

    snowflake.penup()
    snowflake.goto(-150, 90)
    snowflake.pendown()

    order = 5  # Nivel de recursión para el copo de nieve
    size = 500  # Tamaño inicial del copo de nieve

    for _ in range(3):
        koch_snowflake(snowflake, order, size)
        snowflake.right(120)

    snowflake.hideturtle()
    turtle.done()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros del intervalo y resolución
a, b = -10, 10
k = 10000  # Resolución

# Definición de la función
def f(x):
    return x**2

# Valores para la gráfica
z = np.linspace(a, b, k)
y = f(z)

# Número de rectángulos para la aproximación
n = 5
x = np.linspace(a, b, n)
delta = (b - a) / n

# Crear figura y ejes
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(10, 6), dpi=100)

# Dibujar ejes coordenados
axs.axhline(0, color='black', linewidth=1)
axs.axvline(0, color='black', linewidth=1)

# Gráfica de la función y el área bajo la curva
axs.plot(z, y, color='red', label='$f(x) = x^2$', linewidth=2)
axs.fill_between(z, y, color='cyan', alpha=0.3, label='Área bajo $f(x)$')

# Dibujar rectángulos inscritos (azul) y circunscritos (verde)
for i in range(n - 1):
    xi, xi1 = x[i], x[i+1]
    
    # Rectángulo inscrito
    mi = f(xi)
    axs.plot([xi, xi1], [mi, mi], color='darkblue', linewidth=1)
    axs.fill_between([xi, xi1], [mi, mi], color='darkblue', alpha=0.2, label='Inscritos' if i == 0 else "")
    axs.vlines([xi, xi1], ymin=0, ymax=mi, color='blue', linestyle='--', linewidth=1)

    # Rectángulo circunscrito
    Mi = f(xi1)
    axs.plot([xi, xi1], [Mi, Mi], color='green', linewidth=1)
    axs.fill_between([xi, xi1], [Mi, Mi], color='green', alpha=0.2, label='Circunscritos' if i == 0 else "")
    axs.vlines([xi, xi1], ymin=0, ymax=Mi, color='green', linestyle='--', linewidth=1)

# Etiquetas y título
axs.set_title("Aproximación del área bajo $f(x) = x^2$ con rectángulos", fontsize=14)
axs.set_xlabel("Eje x")
axs.set_ylabel("Eje y")

# Leyenda y cuadrícula
axs.legend(loc="upper left")
axs.grid(True, linestyle="--", alpha=0.5)

# Personalización de los ticks
if n <= 14:
    etiquetas = [f"$x_{i}$" for i in range(n)]
    axs.set_xticks(x)
    axs.set_xticklabels(etiquetas)
else:
    axs.set_xticks([a, b])
    axs.set_xticklabels([f"$x_0$", f"$x_{n}$"])

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Paso 1: Definir la función f(x) --- #
def f(x):
    return -x**2  # Puedes cambiar esta función por otra convexa o no convexa

# --- Paso 2: Generar puntos en el intervalo --- #
a, b = -5, 5               # Intervalo
n = 100                    # Número de puntos
x_vals = np.linspace(a, b, n)
y_vals = f(x_vals)

# --- Paso 3: Verificar si es convexa (discretamente) --- #
def es_convexa(x, y):
    for i in range(1, len(x) - 1):
        if y[i] > (y[i-1] + y[i+1]) / 2:
            return False
    return True

# Evaluación de convexidad
convexa = es_convexa(x_vals, y_vals)
print("¿La función es convexa?", "Sí ✅" if convexa else "No ❌")

# --- Paso 4: Graficar la función --- #
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.title("Gráfica de f(x) - Convexa" if convexa else "Gráfica de f(x) - No Convexa")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Definir la función --- #
def f(x):
    return x**2  # Puedes cambiar a otra función

# --- Dominio e intervalo --- #
a, b = -5, 5
x_vals = np.linspace(a, b, 300)
y_vals = f(x_vals)

# --- Verificación formal de convexidad --- #
def es_convexa_formal(f, a, b, muestras=100):
    np.random.seed(0)
    for _ in range(muestras):
        x1 = np.random.uniform(a, b)
        x2 = np.random.uniform(a, b)
        for lam in np.linspace(0, 1, 10):
            x_lambda = lam * x1 + (1 - lam) * x2
            izquierda = f(x_lambda)
            derecha = lam * f(x1) + (1 - lam) * f(x2)
            if izquierda > derecha + 1e-6:  # tolerancia numérica
                print(f"Violación: f({x_lambda:.2f}) = {izquierda:.4f} > {derecha:.4f}")
                return False
    return True

convexa = es_convexa_formal(f, a, b)
print("¿La función es convexa (def. formal)?", "Sí ✅" if convexa else "No ❌")

# --- Graficar función --- #
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x)", color='navy', linewidth=2)

# --- Dibujar 10 rectas secantes --- #
np.random.seed(1)
for _ in range(10):
    x1, x2 = np.random.uniform(a, b, 2)
    while abs(x1 - x2) < 1:
        x1, x2 = np.random.uniform(a, b, 2)
    x1, x2 = sorted([x1, x2])
    y1, y2 = f(x1), f(x2)
    xs = np.linspace(x1, x2, 100)
    ys = y1 + (y2 - y1) / (x2 - x1) * (xs - x1)
    plt.plot(xs, ys, color='red', linestyle='--', linewidth=2, alpha=0.7)
    plt.plot([x1, x2], [y1, y2], 'ro')

# --- Decoración --- #
plt.title("f(x) con Rectas Secantes\nUsando Definición Formal de Convexidad")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend(["f(x)"])
plt.tight_layout()
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random

# Parámetros del problema
a = 0
b = 2
n = 6

# Definimos la función f(x)
def f(x):
    return math.sqrt(x)

# Generamos los puntos de la partición uniforme
x_partition = np.linspace(a, b, n + 1)  # x_0, x_1, ..., x_n

# Lista de puntos aleatorios x_i^* en cada subintervalo (x_{i-1}, x_i)
x_stars = [random.uniform(x_partition[i], x_partition[i + 1]) for i in range(n)]
heights = [f(xi_star) for xi_star in x_stars]  # Alturas de los rectángulos

# Calcular áreas y guardar extremos
areas = [(x_partition[i+1] - x_partition[i]) * heights[i] for i in range(n)]

# Crear figura
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar función f(x)
x_vals = np.linspace(a, b, 500)
y_vals = [f(x) for x in x_vals]
ax.plot(x_vals, y_vals, color='black', label='$f(x) = \\sqrt{x}$')

# Sombrear área bajo la curva
ax.fill_between(x_vals, y_vals, color='skyblue', alpha=0.4)

# Dibujar los rectángulos
for i in range(n):
    xi_left = x_partition[i]
    xi_right = x_partition[i+1]
    height = heights[i]
    
    # Rectángulo como línea horizontal azul
    ax.hlines(height, xi_left, xi_right, colors='blue', linewidth=3)
    
    # Borde izquierdo y derecho como líneas punteadas grises
    ax.vlines([xi_left, xi_right], ymin=0, ymax=height, colors='gray', linestyles='dotted')

# Agregar líneas punteadas en todos los puntos de la partición
for x in x_partition:
    ax.axvline(x, color='gray', linestyle='dotted', linewidth=1)

# Ejes y etiquetas
ax.axhline(0, color='black', linewidth=1.2)
ax.axvline(0, color='black', linewidth=1.2)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("Sumas de Riemann", fontsize=14)
ax.legend()

# Mostrar gráfico
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Mostrar valor aproximado de la suma de Riemann
print("Suma de Riemann aproximada:", sum(areas))
```

# Método Montecarlo para aproximar $\pi$
El método de Montecarlo es una técnica numérica utilizada para resolver problemas mediante la generación de números aleatorios. Recibe su nombre en honor al famoso casino de Montecarlo en Mónaco, que es conocido por los juegos de azar.
El procedimiento básico del método de Montecarlo implica los siguientes pasos:

* Definir un problema matemático o una pregunta que se desea responder.
  
* Crear un modelo o una simulación del problema. 

* Generar un gran número de muestras aleatorias. Estas muestras representan diferentes escenarios o resultados posibles del problema.

* Aplicar el modelo o la simulación a cada muestra y registrar los resultados.

* Utilizar los resultados de las muestras para estimar o resolver el problema original. Esto puede implicar cálculos estadísticos, promedios, intervalos de confianza u otros métodos de análisis.


La idea clave detrás del método de Montecarlo es que al generar un gran número de muestras aleatorias y aplicar el modelo o la simulación a cada una de ellas, se obtiene una aproximación de la solución o la respuesta deseada. Cuanto mayor sea el número de muestras utilizadas, mayor será la precisión de la estimación.

El método de Montecarlo es especialmente útil en problemas complejos o en aquellos que no se pueden resolver de manera analítica o determinista. Permite abordar problemas que involucran incertidumbre, variabilidad o múltiples variables y proporciona una manera de obtener resultados aproximados cuando no es posible obtener una solución exacta.

Es importante destacar que el método de Montecarlo no garantiza una solución exacta, pero puede proporcionar estimaciones precisas con una suficiente cantidad de muestras aleatorias.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed()

lado = 50 
num_granos = 100

plt.figure(figsize=(8, 8))
plt.axis("equal")
plt.xlim(-lado/2, lado/2)
plt.ylim(-lado/2, lado/2)

xs_in, ys_in = [], []
xs_out, ys_out = [], []

for g in range(num_granos):
    x = np.random.uniform(-1, 1) * 0.5 * lado
    y = np.random.uniform(-1, 1) * 0.5 * lado
    if x**2 + y**2 <= (lado/2)**2:
        xs_in.append(x)
        ys_in.append(y)
    else:
        xs_out.append(x)
        ys_out.append(y)

# Usamos plot para simular scatter
plt.plot(xs_in, ys_in, 'ro', markersize=2)   # 'r' = red, 'o' = marker circular
plt.plot(xs_out, ys_out, 'bo', markersize=2) # 'b' = blue

# Dibujo del círculo usando puntos
theta = np.linspace(0, 2 * np.pi, 300)
x_circle = (lado / 2) * np.cos(theta)
y_circle = (lado / 2) * np.sin(theta)
plt.plot(x_circle, y_circle, 'g-', linewidth=2.5)

plt.title("Un experimento con 100 granos de arena esparcidos", fontsize=15)
plt.show()
```
