```python
import matplotlib.pyplot as plt
import numpy as np
import random

a =0;b = 10;k = 100000

def fun(x):
    valor = 'np.sqrt(x)'
    return eval(valor)

x = np.linspace(a,b,k)
y = fun(x)

plt.axhline(0,linestyle = '-', linewidth = 1, color = 'black')
plt.axvline(0,linestyle = '-', linewidth = 1, color = 'black')
#plt.vlines(b, 0, fun(b),color = 'red', linewidth=1.5 )
plt.vlines(a, 0, fun(a),color = 'red', linewidth=1.5 )
#plt.fill_between(x,y, color = 'lightblue', alpha = 0.5 )
plt.plot(x,y, color = 'red', linewidth=1.5)

n = 10
p  = set([random.uniform(a,b) for i in range(n-2)])
while len(p)<n-2:
    p  = set([random.uniform(a,b) for i in range(n-2)])
p = sorted(list(p.union({a},{b})))
m = 100 
sum = 0
for i in range(len(p) -1):
    z = fun(random.uniform(p[i],p[i+1]))
    sum += z*(p[i+1] - p[i])
    plt.plot([p[i],p[i+1]], [z,z], color = 'darkblue', linewidth = 0.5)
    plt.fill_between([p[i],p[i+1]], [z,z], color = 'lightblue', alpha = 0.5 )
    plt.vlines([p[i],p[i+1]], ymin = min(0,z), ymax=max(0,z), linestyle = '--', color= 'b', linewidth=0.5 )
    plt.plot(p[i], z, marker='o', color='blue', markerfacecolor='white', linewidth=1)
    plt.plot(p[i+1], z, marker='o', color='blue', markerfacecolor='white', linewidth=1)  

plt.xlabel('Eje Y')
plt.ylabel('Eje Y')    
plt.title('Sumas de Riemann para una particion aleatoria')
plt.show()

print('La suma de Riemann es:',sum)

```
