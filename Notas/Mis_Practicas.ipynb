```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.array([-3,-1,0,1, 2,3, 4, 5])
y = np.array([1,-5, 3, 10, 8, 1, 5])

# Crear la figura
plt.figure(figsize=(10, 5))

# Ejes coordenados (eje x y eje y en el origen)
plt.axhline(0, color='black', linewidth=1)  # Eje x
plt.axvline(0, color='black', linewidth=1)  # Eje y

# Graficar la funci√≥n escalonada y l√≠neas verticales punteadas
for i in range(len(x) - 1):
    # Tramo horizontal
    plt.plot([x[i], x[i+1]], [y[i], y[i]], color='blue', linewidth=2, label='Escal√≥n' if i == 0 else "")
    
    # Relleno entre el eje x y el valor de la funci√≥n
    plt.fill_between([x[i], x[i+1]], [y[i], y[i]], color='lightblue', alpha=0.4)
    
    # L√≠neas verticales punteadas en extremos del subintervalo
    plt.vlines(x[i], ymin=0, ymax=y[i], linestyle='--', color='gray', linewidth=1)
    plt.vlines(x[i+1], ymin=0, ymax=y[i], linestyle='--', color='gray', linewidth=1)
    
    # Puntos abiertos en los extremos
    plt.plot(x[i], y[i], marker='o', color='blue', markerfacecolor='blue')
    plt.plot(x[i+1], y[i], marker='o', color='blue', markerfacecolor='white')

plt.plot(x[-1], y[-1], marker='o', color='blue', markerfacecolor='blue')


# Configuraci√≥n del gr√°fico
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Funci√≥n escalonada')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Mostrar
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math 

np.random.seed(10)

a = -5; b = 5
k=1000
n=10

def f(x):
    fun = 'np.sin(x)**2'
    return eval(fun)

z = np.linspace(a,b,k)
y = f(z)

plt.axhline(0, linestyle='-', linewidth=1, color='black')
plt.axvline(0, linestyle='-', linewidth=1, color='black')

plt.plot(z,y, color = 'Tomato', linewidth = 1)

x = np.linspace(a,b,n)
sum = 0;
delta = (b-a)/n
k=10
for i  in range(len(x)-1):
    z=np.linspace(x[i],x[i+1],k)
    
    ci = max(np.array([f(w) for w in z]))
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'darkblue', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='blue', alpha=.5)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1 )
    sum += ci*delta 
    
plt.xlabel('x')
plt.ylabel('f(x)=x^2')
plt.title('Suma de Riemann')
plt.show()


print('La suma de Riemann es:',sum)

```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#np.random.seed(10)

a = 0; b = 5 #Aqu√≠ se define el intervalo

k=10000  #N√∫mero de puntos a gr√°ficar
n=8 #N√∫mero de puntos en la particion del intervalo [a,b]

# Enseguida se define la funci√≥n f(x) mediante su regla de correspondencia:
def f(x):
    fun = 'x**2'
    return eval(fun)

z = np.linspace(a,b,k) #Array con los puntos que se van a utilizar para gr√°ficar
y = f(z) #Arrar con las imagenes de los puntos en z

#Se "dibujan" los ejes
plt.axhline(0, linestyle='-', linewidth=1, color='black')
plt.axvline(0, linestyle='-', linewidth=1, color='black')

#Se dibuja la grafica de la funci√≥n f(x)  y se sombrea el area debajo de la gr√°fica
plt.fill_between(z,y,color='gray', alpha=0.5)
plt.plot(z,y, color = 'red', linewidth = 1.5)


x = np.linspace(a,b,n)  #Array que determina la partici√≥n homegenea de [a,b] de longitud n
delta = (b-a)/n #Longitud de cada uno de los subintervalos determinados por dos puntos consecutivos de la partici√≥n homogenea

#Enseguida inicializamos la suma inferior y superior en Cero
suma_inferior = 0;
suma_superior = 0;

#Graficamos los rectangulos inscritos
for i  in range(len(x)-1):
    ci = f(x[i])
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'darkblue', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='darkblue', alpha=0.2)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1.5 )
    suma_inferior += ci*delta

#Graficamos los rectangulos inscritos
for i  in range(len(x)-1):
    ci = f(x[i+1])
    plt.plot([x[i],x[i+1]], [ci,ci], color = 'green', linewidth = 1)
    plt.fill_between([x[i],x[i+1]], [ci,ci], color='green', alpha=0.15)
    plt.vlines([x[i],x[i+1]], ymin = min(0,ci), ymax= max(0,ci), linestyle = '--',color = 'w', linewidth =1.5 )
    suma_superior += ci*delta

plt.xlabel('$x$')
plt.ylabel('$f(x)=x^2$')
#plt.title('Suma de Riemann')
plt.show()


print(f"La suma inferior de Riemann es: {suma_inferior}. ", end='')
print(' |  La suma superior de Riemann es:',suma_superior)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

# Par√°metros
a, b = 0, 5
n = 10

# Funci√≥n continua en [a, b]
def f(x):
    return math.sqrt(x)

# Generar partici√≥n aleatoria ordenada con puntos en (a,b)
internal_points = sorted(random.uniform(a, b) for _ in range(n - 1))
x_partition = [a] + internal_points + [b]

# Generar puntos x_i^* aleatorios dentro de cada subintervalo (x_{i-1}, x_i)
x_stars = [random.uniform(x_partition[i], x_partition[i + 1]) for i in range(n)]
heights = [f(x_star) for x_star in x_stars]

# Crear figura
plt.figure(figsize=(10, 8))
sum = 0
# Graficar los rect√°ngulos estilo escalonado
for i in range(n):
    xi_left, xi_right = x_partition[i], x_partition[i + 1]
    yi = heights[i]
    sum = sum + (xi_right - xi_left)* yi
    # Tramo horizontal azul grueso
    plt.plot([xi_left, xi_right], [yi, yi], color='blue', linewidth=2)

    # Relleno celeste semitransparente
    plt.fill_between([xi_left, xi_right], [0, 0], [yi, yi], color='lightblue', alpha=0.4)

    # L√≠neas verticales punteadas grises
    plt.vlines([xi_left, xi_right], ymin=0, ymax=yi, linestyle='--', color='gray', linewidth=1)

    # Puntos extremos (c√≠rculos blancos con borde azul)
    plt.plot([xi_left, xi_right], [yi, yi], 'o', markerfacecolor='white', markeredgecolor='blue', markersize=4.5)

# Graficar la curva f(x)
x_vals = np.linspace(a, b, 500)
y_vals = [f(x) for x in x_vals]
plt.plot(x_vals, y_vals, color='black')

# Ejes y detalles
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Suma de Riemann de una partici√≥n aleatoria')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print('La suma de Riemann es:',sum)

```
